```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set()
```

```python
#Load the Data
red_wine_original = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',sep=';')
white_wine_original = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',sep=';')
red_wine = red_wine_original.copy()
white_wine = white_wine_original.copy()
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
red_wine

```

```python
# here we are transforming these labels into categrical data type (specific to pandas) instead of simple string
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'],
categories=['low', 'medium', 'high'])
red_wine
```

```python
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
white_wine
```

```python
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'],
categories=['low', 'medium', 'high'])
white_wine
```

```python
#Pie chart for comparing quantities

count = [1599,4898]
wines = ['Red wine', 'white wine']
# Plotting data on chart
plt.pie(count, labels=wines, autopct='%.0f%%')

# Add title to the chart
plt.title('Comparison for quantities of White and Red wine')

# Displaying chart
plt.show() 
```

```python
#Plotting data as a histplot for Red wine dataset

sns.histplot(data = red_wine, x= "quality_label",color='red', bins=10)
plt.xlabel('Distribution of Quality')
plt.ylabel('Count')
plt.ylim(0, 1000)
```

```python
#Plotting data as a histplot for White wine dataset

sns.histplot(data = white_wine, x="quality_label",color='grey', bins=20)
plt.xlabel('Distribution of Quality')
plt.ylabel('Count')
plt.ylim(0, 3500)
```

```python
# Univariate Plotting of Red wine 
red_wine.hist(bins=15, color='red', edgecolor='black', linewidth=1.0, xlabelsize=8, ylabelsize=8, grid=False)

plt.tight_layout(rect=(0, 0, 1.2, 1.2))

rt = plt.suptitle('Red Wine Univariate Plots', x=0.65, y=1.25, fontsize=15)
```

```python
# Univariate Plotting of white wine 

white_wine.hist(bins=15, color='grey', edgecolor='black', linewidth=1.0, xlabelsize=8, ylabelsize=8, grid=False)

plt.tight_layout(rect=(0, 0, 1.2, 1.2))

rt = plt.suptitle('White Wine Univariate Plots', x=0.65, y=1.25, fontsize=15)
```

```python

 #Boxplot for quality label and alcohol
sns.boxplot(x="quality_label",
                y="alcohol",
                data=red_wine)

plt.xlabel('Wine Quality Class')
plt.ylabel('Wine Alcohol %')
```

```python
#Boxplot for quality label and pH
sns.boxplot(x="quality_label",
                y="pH",
                data=red_wine)

plt.xlabel('Wine Quality Class')
plt.ylabel('Wine pH')
```

```python
# Basic heat map to show correlation between features (Red wine) 

plt.figure(figsize=(12,8))

data = red_wine_original.corr()

mask = np.triu(np.ones_like(data.corr()))

sns.heatmap(data, annot=True, cmap='Spectral', vmax=1, vmin=-1, mask = mask)

plt.show()
```

```python

# Basic heat map to show correlation between features (white wine) 

plt.figure(figsize=(12,8))

data = white_wine_original.corr()

mask = np.triu(np.ones_like(data.corr()))

sns.heatmap(data, annot=True, cmap='Spectral', vmax=1, vmin=-1, mask = mask)
#sns.set(rc = {'figure.figsize':(20,10)})

plt.show()
```
